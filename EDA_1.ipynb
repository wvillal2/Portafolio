# **Histogramas**
Se generó histogramas de todas las variables numericas donde:


*   Distribución de lluvia
*   Fertilizante

*   Rendimientos

**Pequeños Resultados**

# **Correlación**

Se seleccionó solo las columnas numericas y se graficó la matriz de correlación

**Resultado:**
- Se observa relación positiva entre fertilizante y rendimiento.
- También hay correlaciones moderadas entre nutrientes (N, P, K) y el rendimiento.
- La lluvia muestra cierta variabilidad, pero no una correlación tan fuerte.


```python
# Seleccionar solo columnas numéricas
numeric_df = df.select_dtypes(include=[np.number])

# Revisar qué columnas quedaron
print(numeric_df.columns)

# Matriz de correlación solo con numéricas
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de correlación (solo numéricas)")
plt.show()
```

# **BoxPlot**
se muestra que el boxplot cambia el rendimiento segun la categoria `temperatue`

```python
print(df.columns)
print(df.dtypes)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Temperatue', y='Yeild (Q/acre)', data=df)
plt.title("Distribución del rendimiento por categoría de Temperatura")
plt.show()
```
